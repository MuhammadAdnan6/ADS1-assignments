## Bar Chart

```python
def bar_chart(dataframe):
    colors = sns.color_palette("Accent", n_colors=6)
    counts = dataframe['Caffeine_Intake'].value_counts().sort_index()    
    plt.figure(figsize=(8, 5))
    sns.barplot(
        x=counts.index, 
        y=counts.values, 
        palette=colors
    )    
    plt.title('Distribution of Caffeine Intake', fontsize=16, pad=20)
    plt.xlabel('Caffeine Intake', fontsize=12)
    plt.ylabel('Count', fontsize=12)
    handles = [
        plt.Rectangle((0, 0), 1, 1, color=colors[i]) 
        for i in range(len(counts.index))
    ]
    labels = [f'Cup Per Day: {int(level)}' for level in counts.index]
    plt.legend(handles, labels, title='Cups Per Day', bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.tight_layout()
    plt.show()
```

```python
bar_chart(dfc)
```

It's obvious from above distribution that mostly students have 2 cups per day which is moderate. Smaller group of students consumes high caffeine or 5 cups per day while notable number of students consumes no caffeine or lower amount of caffeine. High Caffeine might result in bad sleeping habits.

## Scatter Graph

```python
def scatter_graph(dataframe):
    plt.figure(figsize=(10, 6))
    Quality_counts = dataframe['Sleep_Quality'].value_counts().reset_index()
    Quality_counts.columns = ['Sleep_Quality', 'Count']    
    sns.scatterplot(data=Quality_counts, x='Sleep_Quality', y='Count', color='Red', s=100, alpha=0.7)
    plt.title('Sleep Quality Distribution', fontsize=16)
    plt.xlabel('Sleep Quality', fontsize=12)
    plt.ylabel('Count', fontsize=12)
    plt.grid(True)
    plt.show()
```

```python
scatter_graph(df)
```

Scatter graph shows two peaks with one is high number of students experiencing poor sleep quality and very small group of students enjoying good sleep of rating 8. It shows that highest concentraton of student is found at very poor slwwp quality. There is notable number of students who are enjoying moderate sleep, with sleep quality ranging from 3 to 10 excluding 8. 

## Violin Plot

```python
def violin_plot(dataframe):
    correlation_df = dataframe[['Weekday_Sleep_Start', 'Weekend_Sleep_Start', 'Weekday_Sleep_End', 'Weekend_Sleep_End']].corr()
    print("Correlation matrix:")
    print(correlation_df)
    df_long = dataframe.melt(value_vars=['Weekday_Sleep_Start', 'Weekend_Sleep_Start', 'Weekday_Sleep_End', 'Weekend_Sleep_End'],
                             var_name='Sleep_Type',
                             value_name='Time')
    plt.figure(figsize=(10, 6))    
    sns.violinplot(
        data=df_long, 
        x='Sleep_Type', 
        y='Time', 
        inner='box',       
        palette='Set2'
    )
    sns.stripplot(
        data=df_long,
        x='Sleep_Type',
        y='Time',
        color='black',   
        alpha=0.5,     
        jitter=True      
    )
    plt.title('Sleep Timing on Weekdays vs Weekends', fontsize=16, pad=15)
    plt.xlabel('Sleep Timing', fontsize=12)
    plt.ylabel('Time (hours)', fontsize=12)
    handles = [plt.Line2D([0], [0], color=sns.color_palette('Set2')[i], lw=4) for i in range(len(df_long['Sleep_Type'].unique()))]
    plt.legend(handles, df_long['Sleep_Type'].unique(), title='Sleep Type', bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0)

    plt.tight_layout()
    plt.show()
```

```python
 violin_plot(dfc)
```

For both "Weekday_Sleep_Start" and "Weekend_Sleep_Start," most students go to bed in the late evening or early night, with distributions peaking between roughly 21:00 and 24:00 hours. There is a slightly wider range of sleep start times on weekends, as indicated by the shape's width. For sleep end times, most students wake up between 6:00 and 10:00 on weekdays and slightly later on weekends, with a broader distribution for "Weekend_Sleep_End," showing that weekend wake-up times vary more than weekday wake-up times.
